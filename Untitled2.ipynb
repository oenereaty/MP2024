## 데이터

```r
years <- c(1962:1978)
high <- c(25.82, 25.35, 24.29, 24.05, 24.89, 25.35, 25.23, 25.06, 27.13, 27.36, 26.65, 27.13, 27.49, 27.08, 27.51, 27.54, 26.21)
low <- c(18.24, 16.50, 20.26, 20.97, 19.43, 19.31, 20.85, 19.54, 20.49, 21.91, 22.51, 18.81, 19.42, 19.10, 18.80, 18.80, 17.57)

amazon_data <- data.frame(years, high, low)
```

```r
# High 와 Year 의 비교
plot(high ~ years, data = amazon_data,
     pch = 19, cex = 0.5,
     xlab = "years", ylab = "high (m)",
     main = "연도에 따른 아마존 강의 최고 수위")

# Low 와 Year 의 비교
plot(low ~ years, data = amazon_data,
     pch = 19, cex = 0.5,
     xlab = "year", ylab = "low (m)",
     main = "연도에 따른 아마존 강의 최저 수위")

# High 와 Low 의 비교
plot(low ~ high, data = amazon_data,
     pch = 19, cex = 0.5,
     xlab = "high (m)", ylab = "low (m)",
     main = "아마존 강의 최저 수위와 최고 수위의 관계")
```

## 회귀모형

```r
# years 에 대한 high 회귀모형
lm_high <- lm(high ~ years, data = amazon_data)
summary(lm_high)

# years 에 대한 low 회귀모형
lm_low <- lm(low ~ years, data = amazon_data)
summary(lm_low)

# low 에 대한 high 회귀모형
lm_high_low <- lm(high ~ low, data = amazon_data)
summary(lm_high_low)
```
